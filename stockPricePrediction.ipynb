```python
#Pandas is a Python library used for working with data sets.
#It has functions for analyzing, cleaning, exploring, and manipulating data.
#The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" 

import pandas as pd
# Importing yahoo finance for retrieving data
import yfinance as yf 
```

```python
tata = yf.Ticker("TATASTEEL.NS")
```

```python
tata = tata.history(period="max")
```

```python
#Panda's DataFrame
tata
```

```python
tata.index
```

```python
tata.index = pd.to_datetime(tata.index)
```

```python
tata.plot.line(y="Close", use_index=True, color="red")
```

```python
#Deleting unwanted columns
del tata["Dividends"]
del tata["Stock Splits"]
```

```python
#Creating a new column in DataFrame "Tomorrow", showing closing price of the upcoming day
tata["Tomorrow"] = tata["Close"].shift(-1)
```

```python
tata
```

```python
#Target column indicates if tomorrow's price is going to me more than today's price
#{astype converts bool into integer for better use in machine learning}
tata["Target"] = (tata["Tomorrow"] > tata["Close"]).astype(int)
```

```python
#Removing all data before 1990
tata = tata.loc["1996-02-01":].copy()
```

```python
tata
```

```python
#################################################################################
#SKLEARN implement machine learning models and statistical modelling.
#The RandomForestClassifier is an ensemble learning method based on the Random Forest algorithm,
#which builds multiple decision trees and combines their predictions to improve accuracy and reduce overfitting.

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

#This line initializes a RandomForestClassifier model with 300 estimators (decision trees) 
#And a minimum number of samples required to split an internal node set to 150. 
#The random_state=1 is used to ensure reproducibility of the results.

#"reproducibility" refers to the ability to obtain the same or consistent results 
#when running the code on the same dataset and with the same initial conditions
#################################################################################
```

```python
#Splitting Data into "TrainingSet" and "TestingSet"
#train contains all data rows except the last 100 rows
#test contains the last 100 rows of data
train = tata.iloc[:-100]
test = tata.iloc[-100:]
```

```python
predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds
```

```python
#converting the above array into a Pandas series
preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
#Calculating Precision Score
precision_score(test["Target"],preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot(color=["blue", "red"])
```

```python
#######
#Creating new data entries,"RatioColumn" and "TrendColumn"
horizons = [2,5,60]#2= 2 Days,5=week,60=3Months
new_predictors = ["Close", "Volume", "Open", "High", "Low"]

for horizon in horizons:
    rolling_averages = tata.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    tata[ratio_column] = tata["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    tata[trend_column] = tata.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
tata
```

```python
#Removing all Rows containing Incomplete Data"NaN"
tata = tata.dropna(subset=tata.columns[tata.columns != "Tomorrow"])
```

```python
tata
```

```python
#Reimplementation Of Model with NewPredictors
model = RandomForestClassifier(n_estimators=300, min_samples_split=150, random_state=1)
train = tata.iloc[:-100]
test = tata.iloc[-100:]

model.fit(train[new_predictors], train["Target"])

from sklearn.metrics import precision_score
preds = model.predict(test[new_predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["Target"],preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot(color=["blue", "red"])
```
